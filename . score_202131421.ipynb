```python
def average():
    s = 0
    n=int(input("과제의 갯수는: "))

    for i in range(1,n+1): 
        a=float(input("%d번째 과제의 점수: "%(i)))
        s+=a
    

    
    mid = int(input("중간고사 성적을 입력해주세요?"))

    fin = int(input("기말고사 성적을 입력해주세요?"))
    
    H = a * 0.4

    M = mid * 0.3

    F = fin * 0.3

    score = H + M + F
    if 100 >= score >= 90:
        grade = "A"
    elif 90 > score >= 80:
        grade = "B"
    elif 80 > score >= 70:
        grade = "C"
    elif 70 > score >= 60:
        grade = "D"
    elif score < 60:
        grade = "F"
    
    
    print("\n총점이 %d점이므로"%s)
    print("과제의 평균은 %.2f점입니다."%(s/n))
    
    print("과제 반영점수(40%)" , H, "중간 반영점수(30%)" , M, "기말 반영점수(30%)" , F)
    print("성적은" + grade + "입니다.")

average()
과제의 갯수는: 15
1번째 과제의 점수: 80
2번째 과제의 점수: 80
3번째 과제의 점수: 80
4번째 과제의 점수: 80
5번째 과제의 점수: 80
6번째 과제의 점수: 80
7번째 과제의 점수: 80
8번째 과제의 점수: 80
9번째 과제의 점수: 80
10번째 과제의 점수: 80
11번째 과제의 점수: 80
12번째 과제의 점수: 80
13번째 과제의 점수: 80
14번째 과제의 점수: 80
15번째 과제의 점수: 80
중간고사 성적을 입력해주세요?80
기말고사 성적을 입력해주세요?80

총점이 1200점이므로
과제의 평균은 80점입니다.
과제 반영점수(40%) 32.0 중간 반영점수(30%) 24.0 기말 반영점수(30%) 24.0
성적은B입니다.
```
